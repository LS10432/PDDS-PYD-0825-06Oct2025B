### Task 1: Loading the Data:
* Load the Titanic dataset into a Pandas DataFrame.

```python
# Import the pandas library and give it an alias 'pd'
import pandas as pd

# Read the Titanic dataset and load it into a DataFrame named 'df'
df = pd.read_csv('/content/titanic.csv')

# Display the entire DataFrame in the output
display(df)
```

### Task 2: Initial Exploration:
* Display the first 5 rows of the dataset to get an overview.
* Provide basic information about the dataset using the info() method.

```python
# Print a separator line
print("="*60)
# Print header for First 5 rows
print("First 5 rows of the DataFrame")
# Print a separator line
print("="*60)

# Print the first 5 rows of the DataFrame
display(df.head())

print("\n")

# Print a newline followed by a separator line
print("\n" + "="*60)
# Print header for Basic information
print("Basic information of the DataFrame")
# Print a separator line
print("="*60)

# Print a basic information of the DataFrame
df.info()
```

### Task 3: Handling Missing Values:
* Identify and count the number of missing values in each column.
* Decide on an appropriate strategy for handling missing values in at least two columns and implement it.

```python
# Code to count the number of rows in the DataFrame
num_rows = df.shape[0]
print(f"The number of rows in the DataFrame is: {num_rows}")

print("\n")

# Print a separator line
print("="*60)
# Print header for Missing values
print("Missing values before handling:")
# Print a separator line
print("="*60)

# Identify and count missing values in each column
print(df.isnull().sum())

# Handle missing values in 'Age' column by replacing missing value with the mean value
df.fillna({"Age": df["Age"].mean()}, inplace=True)

# Handle missing values in 'Cabin' column by replacing missing value with the mode value
df.fillna({"Cabin": df["Cabin"].mode()[0]}, inplace=True)

# Handle missing values in 'Embarked' column by replacing missing value with the mode value
df.fillna({"Embarked": df["Embarked"].mode()[0]}, inplace=True)

# Print a newline followed by a separator line
print("\n"+"="*60)
# Print header for Missing values
print("Missing values after handling:")
# Print a separator line
print("="*60)

# Identify and count missing values in each column
print(df.isnull().sum())

# Code to count the number of rows in the DataFrame
num_rows = df.shape[0]
print("\n")
print(f"The number of rows in the DataFrame is: {num_rows}")
```

### Task 4: Descriptive Statistics:
* Generate descriptive statistics for the numerical columns using the describe() method.

```python
# Print a separator line
print("="*60)
# Print header for Descriptive statistics
print("Descriptive statistics")
# Print a separator line
print("="*60)

# Generate descriptive statistics for numerical columns
display(df.describe())
```

### Task 5: Unique Values:
* List unique values in the 'Pclass' column.
* Count the occurrences of each unique value in the 'Embarked' column.

```python
# Print the unique values in the 'Pclass' column
print("Unique values in 'Pclass' column:")
display(df['Pclass'].unique())

# Print a newline followed by a separator line
print("\n"+"="*60)
# Print header for Occurrences
print("Occurrences of unique values in 'Embarked' column")
# Print a separator line
print("="*60)

# Count the occurrences of each unique value in the 'Embarked' column
display(df['Embarked'].value_counts())
```

### Task 6: Data Manipulation:
* Create a new column named 'FamilySize' that represents the sum of 'SibSp' and 'Parch'.
* Drop the 'Cabin' column from the dataset.

```python
# Create a new column named 'FamilySize' that represents the sum of 'SibSp' and 'Parch'
df['FamilySize'] = df['SibSp'] + df['Parch']

# Drop the 'Cabin' column from the dataset
df = df.drop('Cabin', axis=1)

# Display the first few rows to show the changes
display(df.head())
```

### Task 7: Filtering and Sorting:
* Create a new DataFrame containing only passengers with an age greater than 30.
* Sort the DataFrame by 'Fare' in descending order.

```python
# Create a new DataFrame containing only passengers with an age greater than 30
df_age_over_30 = df[df['Age'] > 30]

# Print a newline followed by a separator line
print("="*60)
# Print header for Occurrences
print("Passengers with an age greater than 30")
# Print a separator line
print("="*60)

# Display the first 5 rows of the filtered DataFrame
display(df_age_over_30.head())

print("\n")

# Sort the DataFrame by 'Fare' in descending order
df_sorted_by_fare = df_age_over_30.sort_values(by='Fare', ascending=False)

# Print a newline followed by a separator line
print("="*60)
# Print header for Occurrences
print("Sort by 'Fare' in descending order")
# Print a separator line
print("="*60)

# Display the first 5 rows of the sorted DataFrame
display(df_sorted_by_fare.head())
```

### Task 8: Data Analysis:
* Calculate the survival rate for different passenger classes ('Pclass').
* Identify and display the passenger with the highest 'Fare'.

```python
# Calculate the survival rate for different passenger classes ('Pclass')
survival_rate_by_pclass = df.groupby('Pclass')['Survived'].mean()

# Print a newline followed by a separator line
print("="*60)
# Print header for Survival rate
print("Survival rate by Passenger Class:")
# Print a separator line
print("="*60)

# Display Survival rate by Passenger Class as percentages with two decimal places for better readability.
display(survival_rate_by_pclass.apply(lambda x: f'{x:.2%}'))

print("\n")

# Identify and display the passenger with the highest 'Fare'
passenger_highest_fare = df.loc[df['Fare'].idxmax()]

# Print a newline followed by a separator line
print("="*60)
# Print header for Passenger with the highest Fare
print("Passenger with the highest Fare:")
# Print a separator line
print("="*60)

# Display Passenger with the highest Fare
display(passenger_highest_fare)
```

### Conclusion:
The analysis of the Titanic dataset revealed several key findings:
- Missing values were encountered in the ‘Age’, ‘Cabin’, and ‘Embarked’ columns. These were addressed by inputting missing ‘Age’ value with the average value using the mean() function, missing ‘Cabin’ and ‘Embarked’ values with the most frequent value using the mode() function, respectively.
- The survival rate exhibited substantial disparities across passenger classes, with first-class passengers demonstrating the highest survival rate at approximately 63%, while third-class passengers exhibited the lowest at around 24%.
- The passenger with the highest fare was identified as being in first class and having survived.
- A new column named ‘FamilySize’ was created by combining ‘SibSp’ and ‘Parch’, thereby providing insights into the size of families onboard.
- The ‘Cabin’ column was dropped due to a substantial number of missing values.
