## Impute missing values
1. Impute missing values in the dataset columns

```python
# Print the number of missing values per column before imputation
print("Number of missing values per column before imputation:")
print(df.isnull().sum())

print("\n")

# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
print (f"Number of columns with numerical value :{len(numerical_cols)}")

# Impute missing values in numerical columns using the mean
for col in numerical_cols:
    if df[col].isnull().any():
        mean_val = df[col].mean()
        df[col] = df[col].fillna(mean_val)

# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print (f"Number of columns with non-numerical value :{len(categorical_cols)}")

# Impute missing values in categorical columns using the mode
for col in categorical_cols:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0] # mode() can return multiple values, take the first
        df[col] = df[col].fillna(mode_val)

# Print the number of remaining missing values per column
print("\nNumber of remaining missing values per column after imputation:")
print(df.isnull().sum())

# Reasoning for imputation methods
print("\nReasoning for imputation methods:")
print("Mean is used for numerical columns because it represents the central tendency of the data and is a common method for imputing missing numerical values when the data is not heavily skewed.")
print("Mode is used for categorical columns because it represents the most frequent category, which is a suitable way to fill missing values in discrete data where the concept of mean is not applicable.")
```

## Remove duplicate rows
2. Remove duplicate rows from the DataFrame.

```python
# Get the number of rows before removing duplicates
rows_before = df.shape[0]
print(f"Number of rows before removing duplicates: {rows_before}")

# Count the number of duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# Count the number of unique duplicate rows
unique_duplicate_rows = df[df.duplicated()].shape[0]
print(f"Number of unique duplicate rows: {unique_duplicate_rows}")

# Remove duplicate rows
df = df.drop_duplicates()

# Get the number of rows after removing duplicates
rows_after = df.shape[0]
print(f"Number of rows after removing duplicates: {rows_after}")
```

## Create new features
3. Create new features if needed (e.g., extract year and month from the 'date' column).

```python
# Extract year and month from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Explain the reasoning
print("\nReasoning for extracting year and month:")
print("Extracting year and month from the date column is crucial for time-series analysis. It allows for grouping data by specific time periods, identifying seasonal trends, and comparing metrics across different years and months. This granularity in time data is essential for understanding patterns and changes in the COVID-19 data over time.")

# Display the head of the DataFrame to show the new columns
print("\n" + "="*60)
print("Display the first 5 rows of the DataFrame with 'year' and 'month' columns.")
print("\n" + "="*60)
display(df.head())
```

## Investigate geographical variations

### Subtask:
Compare COVID-19 metrics and related factors (e.g., population density, GDP per capita) across different continents and countries using visualizations like bar plots and scatter plots.


**Reasoning**:
Group the DataFrame by continent and calculate the mean for numerical columns, then reset the index and create bar plots to compare average total cases and deaths per million across continents.



```python
import matplotlib.pyplot as plt

# Group by continent and calculate the mean for numerical columns
df_continent_agg = df.groupby('continent').mean(numeric_only=True)

# Reset the index
df_continent_agg = df_continent_agg.reset_index()

# Create a bar plot for average total cases per million by continent
plt.figure(figsize=(12, 6))
plt.bar(df_continent_agg['continent'], df_continent_agg['total_cases_per_million'])
plt.xlabel('Continent')
plt.ylabel('Average Total Cases per Million')
plt.title('Average Total Cases per Million by Continent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create a bar plot for average total deaths per million by continent
plt.figure(figsize=(12, 6))
plt.bar(df_continent_agg['continent'], df_continent_agg['total_deaths_per_million'])
plt.xlabel('Continent')
plt.ylabel('Average Total Deaths per Million')
plt.title('Average Total Deaths per Million by Continent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Group the DataFrame by location and calculate the mean for numerical columns, then reset the index and create bar plots to compare average total cases and deaths per million for the top 10 countries.



```python
# Group by location and calculate the mean for numerical columns
df_country_agg = df.groupby('location').mean(numeric_only=True)

# Reset the index
df_country_agg = df_country_agg.reset_index()

# Sort by total_cases_per_million in descending order and select the top 10
df_top_10_cases = df_country_agg.sort_values(by='total_cases_per_million', ascending=False).head(10)

# Create a bar plot for average total cases per million for the top 10 countries
plt.figure(figsize=(14, 7))
plt.bar(df_top_10_cases['location'], df_top_10_cases['total_cases_per_million'])
plt.xlabel('Country')
plt.ylabel('Average Total Cases per Million')
plt.title('Top 10 Countries by Average Total Cases per Million')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Sort by total_deaths_per_million in descending order and select the top 10
df_top_10_deaths = df_country_agg.sort_values(by='total_deaths_per_million', ascending=False).head(10)

# Create a bar plot for average total deaths per million for the top 10 countries
plt.figure(figsize=(14, 7))
plt.bar(df_top_10_deaths['location'], df_top_10_deaths['total_deaths_per_million'])
plt.xlabel('Country')
plt.ylabel('Average Total Deaths per Million')
plt.title('Top 10 Countries by Average Total Deaths per Million')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Create scatter plots to visualize the relationship between GDP per capita and total cases per million, and population density and total cases per million, using the aggregated country data.



```python
# Create a scatter plot for total cases per million vs. GDP per capita
plt.figure(figsize=(10, 6))
plt.scatter(df_country_agg['gdp_per_capita'], df_country_agg['total_cases_per_million'], alpha=0.5)
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases per Million')
plt.title('Total Cases per Million vs. GDP per Capita by Country')
plt.grid(True)
plt.show()

# Create a scatter plot for total cases per million vs. population density
plt.figure(figsize=(10, 6))
plt.scatter(df_country_agg['population_density'], df_country_agg['total_cases_per_million'], alpha=0.5)
plt.xlabel('Population Density')
plt.ylabel('Total Cases per Million')
plt.title('Total Cases per Million vs. Population Density by Country')
plt.grid(True)
plt.show()
```
