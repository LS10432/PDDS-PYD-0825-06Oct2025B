```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the local path and the URL for the covid dataset
local_path = '/content/owid-covid-data.csv'
covid_url = 'https://lms.educlaas.com/assets/courseware/v1/e206258dd748e86b8bb35fc039e90fa8/asset-v1:EduCLaSS+PYD+PDDS-PYD-0825-06Oct2025B+type@asset+block/owid-covid-data.csv'

try:
    # Try to read the Titanic dataset from the local path
    df = pd.read_csv(local_path)
    print(f"Successfully loaded data from {local_path}")
    print("\n")
except FileNotFoundError:
    # If the local file is not found, read the dataset from the URL
    df = pd.read_csv(covid_url)
    print(f"Local file not found. Loading data from {covid_url}")
    print("\n")

print(f"Total number of columns: {df.shape[1]}")
print(f"Total number of rows: {df.shape[0]}")

# Print a newline followed by a separator line
print("\n" + "="*60)
print("Display the first 5 rows of the DataFrame.")
# Print a newline followed by a separator line
print("\n" + "="*60)
# Print the first 5 rows of the DataFrame
display(df.head())

# Print a newline followed by a separator line
print("\n" + "="*60)
print("Display the last 5 rows of the DataFrame.")
# Print a newline followed by a separator line
print("\n" + "="*60)
# Print the last 5 rows of the DataFrame
display(df.tail())

print("\n" + "="*60)
print("Identify columns with missing values:\n")

# Identify and count missing values in each column
print(df.isnull().sum())
# Print a newline followed by a separator line
print("\n" + "="*60)

# Calculate columns no missing values
missing_cols = df.isnull().sum() != 0

# Display columns no missing values
print(f"Number of columns with no missing values: {len(missing_cols) - missing_cols.sum()}")
print("Column Names with no missing values:")
display(missing_cols[missing_cols == False].index.tolist())

# Print columns missing values
print("\n" + "="*60)
print(f"Number of columns with missing values: {missing_cols.sum()}")

# Calculate percentage of columns with more than 90% missing values.
missing_values_percentage = df.isnull().mean() * 100
columns_to_drop = missing_values_percentage[missing_values_percentage > 90].index.tolist()

# Display columns with more than 90% missing values
print(f"Number of columns with more than 90% missing values: {len(columns_to_drop)}")
print("Column Names with more than 90% missing values:")
print(columns_to_drop)
print("\n" + "="*60)

print("\n"+"="*60)
# Print number of columns to be dropped
print(f"Number of columns to be dropped: {len(columns_to_drop)}")
print("Column Names to be dropped:")
print(columns_to_drop)
print("\n"+"="*60)

# Remove columns with more than 90% missing values
df = df.drop(columns=columns_to_drop)

# Print a newline
print("Display the columns info after drop:")
print("\n")

# Display the columns after drop
print(df.info())
#display(df.columns)

# Print number of columns remaining
print("\n"+"="*60)
print(f"Number of columns after removal: {df.shape[1]}")
print("\n"+"="*60)

# Print a newline followed by a separator line
print("\n" + "="*60)
# Display the data type of the 'date' column
print(f"Data type of 'date' column before conversion: {df['date'].dtype}")

# Print a newline followed by a separator line
print("\n" + "="*60)
print("Column info before conversion:")
print("\n")
# Display summary of the DataFrame
df.info()

# Convert the 'date' column from object type to datetime type
df['date'] = pd.to_datetime(df['date'])

# Print a newline followed by a separator line
print("\n" + "="*60)
# Display the updated data type of the 'date' column
print(f"Data type of 'date' column after conversion: {df['date'].dtype}")

# Print a newline followed by a separator line
print("\n" + "="*60)
print("Column info after conversion:")
print("\n")
# Display summary of the DataFrame
df.info()

print("\n" + "="*60)

# Identify and count missing values in each column
missing_values_per_column = df.isnull().sum()

# Calculate columns with missing values
missing_cols_count = missing_values_per_column[missing_values_per_column != 0].count()

# Calculate numerical and non-numerical columns with missing values
numerical_cols_with_missing = df.select_dtypes(include=['float64', 'int64']).columns[df.select_dtypes(include=['float64', 'int64']).isnull().any()]
categorical_cols_with_missing = df.select_dtypes(include=['object']).columns[df.select_dtypes(include=['object']).isnull().any()]

# Display number of columns missing values and numerical and non-numerical columns with missing values
print(f"Total number of columns: {df.shape[1]}")
print(f"Number of columns with missing values: {missing_cols_count}")
print(f"Number of numerical columns with missing values: {len(numerical_cols_with_missing)}")
print(f"Number of non-numerical columns with missing values: {len(categorical_cols_with_missing)}")
print("\n" + "="*60)

# Select numerical columns
print(f"Number of numerical columns with missing values: {len(numerical_cols_with_missing)}")
print("Use the built-in mean() function to replace missing values in numerical columns. Mean is a common method for imputing missing numerical values when the data is not heavily skewed, as it represents the central tendency of the data.")

# Impute missing values in numerical columns using the mean
for col in numerical_cols_with_missing:
  mean_val = df[col].mean()
  df[col] = df[col].fillna(mean_val)

# Print the number of remaining missing values per column fter numerical imputation
print("\nNumber of remaining missing values per column after numerical imputation:")
print(df.isnull().sum())
print("\n" + "="*60)

# Select categorical columns
print(f"Number of non-numerical columns with missing values: {len(categorical_cols_with_missing)}")
print("Use the built-in mode() function to replace missing values in non-numerical columns. Mode is suitable for categorical columns as it represents the most frequent category, making it ideal for filling missing values in discrete data where mean is not applicable.")

# Impute missing values in categorical columns using the mode
for col in categorical_cols_with_missing:
  mode_val = df[col].mode()[0] # mode() can return multiple values, take the first
  df[col] = df[col].fillna(mode_val)

# Print the number of remaining missing values per column
print("\nNumber of remaining missing values per column after non-numerical imputation:")
print(df.isnull().sum())

print("\n" + "="*60)

# Get the number of rows before removing duplicates
rows_before = df.shape[0]
print(f"Number of rows before removing duplicates: {rows_before}")

# Count the number of duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# Remove duplicate rows
df = df.drop_duplicates()

# Get the number of rows after removing duplicates
rows_after = df.shape[0]
print(f"Number of rows after removing duplicates: {rows_after}")

# Extract year and month from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Display the head of the DataFrame to show the new columns
print("\n" + "="*60)
print("Display the first 5 rows of the DataFrame with 'year' and 'month' columns.")
print("\n" + "="*60)
display(df.head())

print("\n" + "="*60)
df.info()

# Get unique countries from the 'location' column
unique_countries = df['location'].unique()

# Count the number of unique countries
num_unique_countries = len(unique_countries)

print("\n" + "="*60)
# Print the total count of unique countries
print(f"Total number of unique countries: {num_unique_countries}")
print("\n" + "="*60)

# Display the list of unique countries
print("\nList of unique countries:")
print(unique_countries)
```
