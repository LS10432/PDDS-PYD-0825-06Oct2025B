### Task 1: Load the Dataset and Explore
* Download the Titanic dataset and assign it to a DataFrame.
* Display the first few rows to get an overview of the data.

```python
# Import the pandas library and give it an alias 'pd'
import pandas as pd

# Define the local path and the URL for the Titanic dataset
local_path = '/content/titanic.csv'
titanic_url = 'https://raw.githubusercontent.com/LS10432/PDDS-PYD-0825-06Oct2025B/refs/heads/main/03/titanic.csv'

try:
    # Try to read the Titanic dataset from the local path
    df = pd.read_csv(local_path)
    print(f"Successfully loaded data from {local_path}")
    print("\n")
except FileNotFoundError:
    # If the local file is not found, read the dataset from the URL
    df = pd.read_csv(titanic_url)
    print(f"Local file not found. Loading data from {titanic_url}")
    print("\n")

# Print the first few rows of the Titanic dataset
display(df.head())

# Print the info of the Titanic dataset
df.info()
```

### Task 2: Visualize Passenger Survival
* Create a bar plot using seaborn to visualize the count of survivors and non-survivors.

```python
# Import the seaborn library and give it an alias 'sns'
import seaborn as sns
# Import the matplotlib library and give it an alias 'plt'
import matplotlib.pyplot as plt

# Create a bar plot for survival count with a width of 6 inches and height of 4 inches
plt.figure(figsize=(6, 4))

# Generate a bar plot showing the count of each unique value in the 'Survived' column using seaborn
sns.countplot(x='Survived', data=df, hue='Survived')

# Set the title of the plot
plt.title('Survival Count (0=Not Survived, 1=Survived)')

# Set the label for the x-axis
plt.xlabel('Survived')

# Set the label for the y-axis
plt.ylabel('Count')

# Replace default x-axis tick labels with more descriptive ones
plt.xticks([0, 1], ['Not Survived', 'Survived'])

# Display the resulting bar plot
plt.show()
```

### Task 3: Analyze Passenger Age Distribution
* Create a histogram using matplotlib to show the distribution of passenger ages.

```python
# Import the matplotlib library and give it an alias 'plt'
import matplotlib.pyplot as plt

# Create a histogram for passenger age distribution with a size of 10 inches by 6 inches
plt.figure(figsize=(10, 6))

# Plot a histogram of the 'Age' column from df, excluding missing values, with 40 bins and black edges
plt.hist(df['Age'].dropna(), bins=40, edgecolor='black')

# Set the title of the plot
plt.title('Distribution of Passenger Ages')

# Label the x-axis as 'Age'
plt.xlabel('Age')

# Label the y-axis as 'Frequency'
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

### Task 4: Explore Class-wise Distribution
* Use seaborn to create a count plot to display the number of passengers in each class.

```python
# Import the seaborn library and give it an alias 'sns'
import seaborn as sns
# Import the matplotlib library and give it an alias 'plt'
import matplotlib.pyplot as plt

# Create a count plot for passenger class distribution with a size of 8 inches by 4 inches
plt.figure(figsize=(8, 4))

# Create a count plot showing the frequency of each passenger class in the 'Pclass' column
sns.countplot(x='Pclass', data=df, hue='Pclass', palette='pastel')

# Set the title of the plot to 'Distribution of Passenger Classes'
plt.title('Distribution of Passenger Classes 1=1st Class, 2=2nd Class, 3=3rd Class')

# Label the x-axis as 'Passenger Class'
plt.xlabel('Passenger Class')

# Label the y-axis as 'Count'
plt.ylabel('Count')

# Set custom x-axis tick labels at positions 0, 1, 2 to '1st Class', '2nd Class', '3rd Class'
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])

# Display the plot
plt.show()
```

### Task 5: Investigate Fare Distribution
* Create a box plot using seaborn to visualize the distribution of passenger fares.

```python
# Import the seaborn library and give it an alias 'sns'
import seaborn as sns
# Import the matplotlib library and give it an alias 'plt'
import matplotlib.pyplot as plt

# Create a box plot for passenger fare distribution with a size of 6 inches by 8 inches
plt.figure(figsize=(6, 8))

# Create a box plot showing the distribution of values in the 'Fare' column vertically
sns.boxplot(x='Pclass', y='Fare', data=df)

# Set the title of the plot to 'Distribution of Passenger Fares'
plt.title('Distribution of Passenger Fares')

# Label the x-axis as 'Pclass'
plt.xlabel('Pclass')

# Label the y-axis as 'Fare'
plt.ylabel('Fare')

# Display the plot
plt.show()
```

### Task 6: Correlation Between Variables
* Generate a heatmap using seaborn to visualize the correlation matrix of the numerical variables in the dataset.

```python
# Import the seaborn library and give it an alias 'sns'
import seaborn as sns
# Import the matplotlib library and give it an alias 'plt'
import matplotlib.pyplot as plt

# Select only columns with numerical data types from df and store in numerical_df
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix showing relationships between all numerical columns
correlation_matrix = numerical_df.corr()

# Create a heatmap with a size of 10 inches by 8 inches
plt.figure(figsize=(10, 8))

# Create a heatmap of the correlation matrix with correlation values displayed, using a coolwarm color scheme, and formatting values to 2 decimal places
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Set the title of the plot to 'Correlation Matrix of Numerical Variables'
plt.title('Correlation Matrix of Numerical Variables')

# Display the plot
plt.show()
```

### Task 7: Visualize Passenger Class and Age
* Create a scatter plot using matplotlib to show the relationship between passenger age and class.

```python
# Import the matplotlib library and give it an alias 'plt'
import matplotlib.pyplot as plt

# Create a scatter plot for passenger age and class with a size of 10 inches by 2 inches
plt.figure(figsize=(10, 2))

# Create a scatter plot with Age on x-axis and Pclass on y-axis, with 50% transparency for overlapping points
plt.scatter(df['Age'], df['Pclass'], alpha=0.5)

# Set the title of the plot to 'Passenger Age vs. Class'
plt.title('Passenger Age vs. Class')

# Label the x-axis as 'Age'
plt.xlabel('Age')

# Label the y-axis as 'Passenger Class'
plt.ylabel('Passenger Class')

# Set custom y-axis tick labels at positions 1, 2, 3 to '1st Class', '2nd Class', '3rd Class'
plt.yticks([1, 2, 3], ['1st Class', '2nd Class', '3rd Class'])

# Enable grid lines on the plot for easier reading
plt.grid(True)

# Display the plot
plt.show()
```
